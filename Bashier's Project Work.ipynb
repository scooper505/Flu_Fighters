```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
```

```python
input_file = "counties.csv"
counties_df = pd.read_csv(input_file)

counties_df.head()
```

```python
states=[]
counties=[]
year=[]
count=[]

target_state=["NJ","RI","MA","CT", "MD", "DE", "NY", "FL", "PA", "OH"]
target_year=[2012,2013,2014,2015,2016,2017]
# https://fluvaccineapi.hhs.gov/api/v2/vaccination_rates/{m}/states/{x}/counties.json?ethnicity=T&medicare_status=A
for m in target_year:
   for x in target_state:
       try:
           data=requests.get(f"https://fluvaccineapi.hhs.gov/api/v2/vaccination_rates/{m}/states/{x}/counties.json?ethnicity=T&medicare_status=A").json()
#         print(json.dumps(data,indent=4))

           for i in range(len(data)) :
               year_order=data[i]["year"]
               year.append(year_order)
               number_county=data[i]["percentage"]
               count.append(number_county)
               state=data[i]["state"]
               states.append(state)
               county=data[i]["name"]
               counties.append(county)

               print(f"the total number of people who get vaccine in {year_order}  for {state} {county} week is {number_county}." )
       except:
           print("....")
```

```python
county_df_percentage = pd.DataFrame({"Year":year,"State":states,"County":counties,"Number of people who got the flu vaccine":count})
county_df_percentage
```

```python
flu_sum_per_state_percentage = county_df_percentage.groupby(["State","Year"]).mean()
flu_sum_per_state_percentage
```

```python
NJ_flu_shots_percentage = flu_sum_per_state_percentage.loc["NJ"]
NJ_flu_shots_percentage

RI_flu_shots_percentage = flu_sum_per_state_percentage.loc["RI"]
RI_flu_shots_percentage

MA_flu_shots_percentage = flu_sum_per_state_percentage.loc["MA"]
MA_flu_shots_percentage

CT_flu_shots_percentage = flu_sum_per_state_percentage.loc["CT"]
CT_flu_shots_percentage    

MD_flu_shots_percentage = flu_sum_per_state_percentage.loc["MD"]
MD_flu_shots_percentage

DE_flu_shots_percentage = flu_sum_per_state_percentage.loc["DE"]
DE_flu_shots_percentage

NY_flu_shots_percentage = flu_sum_per_state_percentage.loc["NY"]
NY_flu_shots_percentage

FL_flu_shots_percentage = flu_sum_per_state_percentage.loc["FL"]
FL_flu_shots_percentage

PA_flu_shots_percentage = flu_sum_per_state_percentage.loc["PA"]
PA_flu_shots_percentage

OH_flu_shots_percentage = flu_sum_per_state_percentage.loc["OH"]
OH_flu_shots_percentage



```

```python
with plt.style.context(('dark_background')):

    Year = [2012,2013,2014,2015,2016,2017]

    fig, ax = plt.subplots(figsize=(8, 6))


    plt.plot(Year, NJ_flu_shots_percentage, label="New Jersey", linewidth=3)
    plt.plot(Year, RI_flu_shots_percentage, label="Rhode Island", linewidth=3)
    plt.plot(Year, MA_flu_shots_percentage, label="Massachussets", linewidth=3)
    plt.plot(Year, CT_flu_shots_percentage, label="Connecticut", linewidth=3)
    plt.plot(Year, MD_flu_shots_percentage, label="Maryland", linewidth=3)
    plt.plot(Year, DE_flu_shots_percentage, label="Deleware", linewidth=3)
    plt.plot(Year, NY_flu_shots_percentage, label="New York", linewidth=3)
    plt.plot(Year, FL_flu_shots_percentage, label="Florida", linewidth=3)
    plt.plot(Year, PA_flu_shots_percentage, label="Pennyslvania", linewidth=3)
    plt.plot(Year, OH_flu_shots_percentage, label="Ohio", linewidth=3)


    plt.xlabel("Year")
    plt.ylabel("Percentage of people who have taken flu shot per state")
    plt.legend(loc="best")
    plt.title("Percentage of flu shots per state")

    plt.show()
```
